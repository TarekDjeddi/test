```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

# Load CSV files into Dataframe

```python
df_members=pd.read_csv('members.csv') # load members table into a dataframe
df_members.head()

df_bookings=pd.read_csv('bookings.csv') # load bookings table into a dataframe
df_bookings.head()

df_facilities=pd.read_csv('facilities.csv') # load facilites table into a dataframe
df_facilities.head()
```

# Retrieve the start times of member named 'David Farrell'

```python
df_book_mem=pd.merge(df_members, df_bookings, how='inner')
df_book_mem_st=df_book_mem[['surname', 'firstname', 'starttime']]
df_st_David=df_book_mem_st['starttime'][(df_book_mem_st['surname']=='Farrell') & (df_book_mem_st['firstname']=='David')]
df_st_David
```

# Work out the start times of bookings for tennis courts

```python
df_book_fac=pd.merge(df_bookings, df_facilities, how='inner')
df_book_fac_st=df_book_fac[['starttime', 'name']]
df_book_fac_tennis=df_book_fac_st[(df_book_fac_st['starttime'] >=  '2012-09-21') & (df_book_fac_st['starttime'] <  '2012-09-22') & ((df_book_fac_st['name'] == 'Tennis Court 1') | (df_book_fac_st['name'] == 'Tennis Court 2'))]
df_book_fac_tennis
```

# Produce a list of all members, along with their recommender

```python
df_mem_rec=df_members[['firstname', 'surname', 'recommendedby']]
df_mem_rec
df_mem_rec["member"] = df_mem_rec["firstname"] + ' ' + df_mem_rec["surname"]
df_mem_rec.drop(['firstname', 'surname'], axis=1, inplace=True)
df_mem_rec=df_mem_rec[['member', 'recommendedby']]
df_mem_rec.head()
```

# List the total slots booked per facility

```python
df_bookings.groupby('facid', sort = True)['slots'].sum('slots')
```

# List the total slots booked per facility in a given month

```python
df_bookings_sep = df_bookings[(df_bookings['starttime'] >=  '2012-09-01') & (df_bookings['starttime'] <  '2012-10-01')]
df_bookings_sep.groupby('facid')['slots'].sum('slots').sort_values(ascending=True)
```

# List the total slots booked per facility per month

```python
df_bookings_month = df_bookings
df_bookings_month['starttime']= pd.to_datetime(df_bookings['starttime'])
df_bookings_month['month'] = pd.DatetimeIndex(df_bookings_month['starttime']).month
df_bookings_month = df_bookings_month[['facid','month','slots']]
df_bookings_month.groupby(['facid', 'month'])['slots'].sum('slots')
```

# List each member's first booking after September 1st 2012

```python
df_book_mem_merge = pd.merge(df_members, df_bookings, how='inner')
df_book_mem_merge = df_book_mem_merge[['surname', 'firstname', 'memid', 'starttime']]

df_book_mem_afterSep = df_book_mem_merge[df_book_mem_merge['starttime'] >= '2012-09-01']

df_book_mem_afterSep = df_book_mem_afterSep.sort_values('starttime', ascending=True)
df_book_mem_afterSep = df_book_mem_afterSep.groupby('memid',as_index=False).agg({'surname':'first', 'firstname':'first', 'starttime':'first'})

df_book_mem_afterSep = df_book_mem_afterSep[['surname', 'firstname', 'memid', 'starttime']]
df_book_mem_afterSep
```
